```python
bios.loc[(bios["height_cm"]>215),["name","height_cm","weight_kg"]]
```

## The & operator (bitwise AND) has higher precedence than comparison operators (>, ==). So, 215 & bios["born_country"] is evaluated first, which leads to an incorrect logical comparison. so alway use () parantheses

## lambda function use 

```python
bios["height_category"]=bios["height_cm"].apply(lambda x: "Short" if x<165 else("Average" if x<185 else "Tall"))
```

```python
bios
```

```python
#we can also define custom function
def categorize_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] < 185 or row['weight_kg'] <= 80:
        return 'Middleweight'
    
    else:
        return 'Heavyweight'
    
bios['Category'] = bios.apply(categorize_athlete, axis=1) #axis=0 is for column and axis=1 is for row
```
