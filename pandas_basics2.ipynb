## merging and concatenating data

```python
nocs=pd.read_csv("https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/noc_regions.csv")
bios=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/bios.csv')
nocs.head()
```

```python
bios.head()
```

```python
bios_new=pd.merge(bios,nocs,left_on="born_country",right_on="NOC",how="left",suffixes=["bios","nocs"])  #left merge 
#suffixes used to handle cases where the two DataFrames being merged have columns with the same names bydefault:_x and _y
bios_new.head()
```

```python
bios_new.rename(columns={"region":"born_country_full"},inplace=True)
```

```python
bios_new.loc[(bios_new["NOCbios"]!=bios_new["born_country_full"]),["name","NOCbios","born_country_full"]]
```

```python
#CONCATENATION
```

```python
france=bios[bios["born_country"]=="FRA"].copy()
rus=bios[bios["born_country"]=="RUS"].copy()
new=pd.concat([france,rus])
```

```python
#Handling null values
```

```python
coffee=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/warmup-data/coffee.csv')
```

```python
coffee.loc[[0,1],["Units Sold"]]=np.nan
```

```python
coffee.head()
```

```python
coffee.isna().sum()
```

```python
coffee.fillna(10000) #changes units sold of 1st and 2nd row to 10000
```

```python
coffee.fillna(coffee['Units Sold'].mean()) #fill nan with mean value of units sold of all coffee type
```

```python
coffee.loc[[0,1],["Units Sold"]]=15
coffee.loc[[2,3],["Units Sold"]]=np.nan
coffee["Units Sold"]=(coffee["Units Sold"].interpolate()) #interpolate assign value by seeing neighbours in table
coffee
```

```python
#coffee.dropna() used to drop any nan row in table
#coffee.dropna(subset=["Unit Sold"]) drop nan row only of Units Sold
#coffee[coffee["Units Sold"].isna()] gives rows where units sold is NAN , .notna() is used to give rows not NAN
```

```python
coffee.head()
```

```python
coffee.groupby(["Coffee Type"])["Units Sold"].sum()
```

```python
coffee.groupby(["Day"])["Units Sold"].sum()
```

```python
coffee["Price"]=15
```

```python
coffee.groupby(["Coffee Type","Day"]).agg({"Units Sold": "sum","Price": "mean"})
```

```python
coffee["revenue"]=coffee["Units Sold"]*coffee["Price"]
```

```python
#Pivot Table
pivot=coffee.pivot(columns=["Coffee Type"],index=['Day'],values=["revenue"])
pivot
```

```python
pivot.sum()
```

## Advance functions

```python
#coffee.drop(0,inplace=True)
coffee["Yesterday_Revenue"]=coffee["revenue"].shift(1)
coffee["perc_increase"]=(coffee["revenue"]/coffee["Yesterday_Revenue"])*100
coffee
```

```python
#RANKING BASED ON SPECIFIC COLUMN
bios["height_rank"]=bios["height_cm"].rank() #this will give smallest height rank 1
bios.sort_values(["height_rank"],ascending=False)
#bios["height_rank"]=bios["height_cm"].rank(ascending=False) this will give tallest height rank 1
```

```python
#Cummulatives Sum
coffee["Cummulative_Revenue"]=coffee["revenue"].cumsum()
coffee
```

```python
#Rolling Function
latte=coffee[coffee["Coffee Type"]=="Latte"].copy()
latte["3day"]=latte["Units Sold"].rolling(3).sum()
latte #3rd row of 3day = sum of first 3 rows of units sold
```
